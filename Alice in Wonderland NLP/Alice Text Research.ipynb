```python
'''
These import statements all load the neccessary packages need to perform the necessary functions to clean and analyze the data. 
'''
import spacy
import nltk
nlp = spacy.load('en')
import textacy
from collections import Counter
from string import punctuation
```

```python
'''
This function is used to count the number of words in any given text file, provided its encoded as utf-8 
and supplied a number of words to count will provide the top N words of that file. I created this function by looking at
stack exchange and customizing a function that was created by user Padraic Cunningham on the question
Print 10 most frequently occurring words of a text that including and excluding stopwords
'''
def content_text(text, number_of_words):
    stopwords = set(nltk.corpus.stopwords.words('english'))
    words = Counter()
    with open(text, encoding='utf-8') as f:
        for line in f:
            split = line.split()
            words.update(w.lower().rstrip(punctuation) for w in split if w not in stopwords)
    return [x for x in words.most_common(number_of_words)]
```

```python
common_alice_words = content_text("alice in winderland.txt", 4)
print(common_alice_words)
```

```python
'''
This cell gets the alice text into a variable called alice and then I apply spacy tokenization on it to get ready to use textacy
on the file
'''
alice = open("alice in winderland.txt", encoding='utf-8')
alice_text = nlp(alice.read())
```

```python
'''
this textacy function extracts ngrams from the text file, so since I wanted proper nouns and adjectives I used a n of two to
create a list of bigrams and to filter them I used the tags for proper noun and adj
'''
alice_word_pairs = textacy.extract.ngrams(alice_text, n=2, filter_nums=True, include_pos=["PROPN","ADJ"])
```

```python

for i in alice_word_pairs:
    print(i)
```
